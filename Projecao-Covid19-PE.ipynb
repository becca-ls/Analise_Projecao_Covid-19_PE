# **Métodos**

## **Modelo SVIRS**

Para explicar e projetar o comportamento da pandemia da Covid-19 em Pernambuco, será utilizado o modelo SVIRS (*Susceptible - Vaccinated - Infectious - Recovered - Susceptible)*. Esse modelo é uma variação do modelo SIR que inclui um novo grupo para descrever a população: os vacinados. Além disso, o modelo considera a possibilidade de reinfecção, ou seja, que a imunidade adquirida ao ser vacinado ou ao se recuperar da doença não é vitalícia e, que por essa razão, um indivíduo pode retornar ao grupo dos suscetíveis e contrair a doença novamente.

No modelo SVIRS, a população total $N$ está dividida entre três grupos:

1.   **Suscetíveis $(S)$:** os indivíduos que podem fazer parte do grupo de infectados.
2.   **Vacinados $(V)$:** os indivíduos que tomaram as duas doses da vacina e que por isso têm menor chance de contrair a doença.
3. **Infectados $(I)$:** os indivíduos com a doença.
4. **Recuperados $(R)$:** os indivíduos que estavam infectados, mas se recuperaram, tendo um tempo de imunidade antes de voltarem a ser suscetíveis novamente. Esta é uma das diferenças do modelo SVIRS para o SIR. No SIR, os recuperados possuem imunidade vitalícia, não podendo se tornar suscetíveis novamente.

Dessa forma, a população total $N$ é descrita como:

  $N=S+ V + I+ R$

Esse modelo é descrito através das seguintes ODEs:

$  \frac{dS}{dt}= \omega_rR -\lambda S + \omega_v V - \phi S $

$  \frac{dV}{dt}=- (1- \psi)\lambda V - \omega_v V + \phi S  $

$ \frac{dI}{dt}= \lambda S + (1-\psi)\lambda V - \gamma I $

$ \frac{dR}{dt}= \gamma I - \omega_r R $

Em que:

1.  $\lambda$ representa a taxa de infecção por pessoa e é dada por $ \lambda= \frac{\beta I}{N}$

2. $\beta$ representa a probabilidade de transmitir a doença entre um indivíduo suscetível e um infectado

3. $\gamma$ representa a taxa de recuperação e é determinado por $\gamma = \frac{1}D$, em que $D$ é a duração média da infecção em dias

4. $\omega_r$ representa a taxa com a qual os indivíduos recuperados voltam para o grupo dos suscetíveis devido à perda de imunidade

5. $\omega_v$ representa a taxa com a qual os indivíduos vacinados voltam para o grupo dos suscetíveis devido à perda de imunidade

6. $\phi$ representa a taxa de vacinação

7. $\psi$ representa a taxa de eficácia da vacina



## **Parâmetros**

Para analisar cada um dos cenários que serão modelados, será necessário ajustar os parâmetros do modelo. Sabemos que as taxas de eficácia da vacina variam conforme sua produção (CoronaVac, AstraZeneca, Pfizer e Janssen), consideraremos aqui, em busca de uma precisão maior, uma média ponderada das eficácias dessas diferentes variações da vacina do coronavírus. Então, a partir do *Acompanhamento do Recebimento de Vacinas para COVID-19* disponibilizado pelo Governo do Estado de Pernambuco e das eficácias da vacina por laboratório,podemos fazer a seguinte análise:

### **Cálculo da Eficácia Média** $ \rightarrow \psi$

*Coronavac* → 4.481.720 doses distribuídas → $P_c = 31,25\%$

> **Eficácia → $\psi_c =   67,4\%$**

*Astrazeneca* → 4.707.170 doses distribuídas → $P_a = 32,83\%$

>**Eficácia → $\psi_a =   76\%$** *(Média aritmética entre 60% e 92%)*

*Pfizer* → 4.977.180 doses distribuídas → $P_p =  34,71\%$

>**Eficácia →  $\psi_p =   83,5\%$** *(Média aritmética entre 77% e 90%)*

*Janssen* → 173.810 doses distribuídas → $P_j =  1,21 \%$

>**Eficácia →  $\psi_j =   66\%$**

 ***Total de doses distribuídas  → 14.339.880***

Onde $P$ é a porcentagem de doses no total de doses e $\psi$ é a eficácia da vacina de determinado laboratório.

Dessa forma, a eficácia média das vacinas é dada por:

$$\psi = \frac{\psi_cP_c + \psi_aP_a +\psi_pP_p +\psi_jP_j}{100\%}$$



$$\psi = \frac{0,674(31,25 \%)+0,76(32,83\%)+0,835(34,71\%)+0,66(1,21\%)}{100\%} =  0,7579475$$



$$\psi ≅ 0,76$$ 

### **Cálculo de $\omega_v$, $\omega_r$ e $\phi$**

Para encontrar as taxas de perda de imunidade (pela vacinação e pela recuperação da doença) e de vacinação, transformamos as taxas anuais ou mensais em taxas diárias.

Estima-se que a imunidade da vacinação $\omega_v$ cai 50% ao ano, então, calculamos que essa taxa é de 0,1170% ao dia.

A imunidade adquirida após a recuperação da doença $\omega_r$cai 10% a cada 90 dias, por isso calculamos que essa taxa é de 0,1897% ao dia.

Para calcular a taxa de vacinação diária $\phi$, consideramos que essa vacinação foi feita uniformemente durante os dias, por isso, sabendo o número de vacinados no dia 31/10/2021 e o dia de início da vacinação, calculamos que essa taxa é de 0,2627% ao dia.

### **Cálculo de $\beta$**

A taxa de transmissão da doença $\beta$ pode ser calculada a partir da taxa de propagação da Covid-19 na população $r$ e do número de pessoas suscetíveis $S$ da seguinte forma:

$\beta = \frac{r}{S}$

A taxa de propagação $r$ foi encontrada a partir da análise das taxas de propagação da doença na realidade considerando as ondas que ocorreram no número de casos ao longo do tempo.

## **Realidade**

Para comparar os resultados do modelo com a realidade, é necessário primeiramente analisar e projetar dados sobre a Covid-19 em um gráfico. Para isso, utilizamos as bases de dados "COVID-19 em Dados" e "vacinados-PE" para exibir o número de casos acumulados e número de vacinados em Pernambuco respectivamente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
dataset = pd.read_csv('COVID-19 em Dados.csv', sep=',')

pe = dataset.fillna('0') # preenche celulas vazias com zero

# convertendo valores para inteiro
pe["confirmados"].astype(str).astype(int)

# selecionando colunas do dataset
df_graph = pd.DataFrame({'Casos acumulados': pe['confirmados'].to_list(), 'Data': pe['dt_referencia'].to_list()})

# selecionando linhas entre o período de 01/07/2020 e 31/10/2021
df_graph['Data'] = pd.to_datetime(df_graph['Data'])
datas = (df_graph['Data'] >= '2020-07-01') & (df_graph['Data'] <= '2021-10-31')
dt_filtrado = df_graph[datas]

# exibindo casos acumulados em um gráfico
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], dt_filtrado['Casos acumulados'],'r', lw=2, label = 'Casos acumulados')
plt.legend()
plt.show()
```

## **Modelo**

Usamos a função Odeint do pacote Scipy da biblioteca científica Python para integrar o sistema de equações diferenciais com o tempo $dt$ = 1,0/1440. Nos casos investigados, consideramos o tempo de recuperação como $D$ = 15 dias. Supomos que esse parâmetro é uniforme e não muda sensivelmente durante a escala de tempo de alguns meses. Além dos parâmetros explicados anteriormente, os valores iniciais utilizados são: 

$N = 9 674 793$

$V_0 = 0$

$I_0 = 14 226$

$R_0 = 41 925$

$S_0 = N - I_0 - R_0 - V_0$

A partir disso, aplicaremos nosso modelo para explicar o comportamento da pandemia e projetar alguns cenários sobre a COVID-19 em Pernambuco.

```python
# DEFININDO AS CONDIÇÕES INICIAIS

N = 9674793
V0, I0, R0 = 0, 14226, 41925
S0 = N - I0 - R0 - V0

omegaV = 0.001170 # imunidade da vacina cai 50% ao ano
omegaR = 0.001897 # imunidade de recuperados cai 10% a cada 90 dias

psi = 0.76 # eficácia da vacina
phi = 0.002627 # taxa de vacinação

D = 15 # infecção dura 15 dias
gamma = 1./D
```

### **Utilizando o Modelo SIR**

Antes de aplicar o modelo SVIRS, iremos aplicar o model SIR para, em seguida, identificar as diferenças entre as curvas encontradas a partir dele e a partir do nosso modelo.

```python
beta, gamma = 0.069/S0, 1./15 

def deriv2(y, t, N, beta, gamma):
    if t < 70: # meio de agosto 2020
      r = 0.069
    elif t < 160: # meio de novembro 2020
      r = 0.06
    elif t < 170: # fim de novembro 2020
      r = 0.13
    elif t < 380: # fim de abril 2021
      r = 0.069
    else:
      r = 0.053

    S, I, R = y
    beta = r/S

    dSdt = -beta*S*I 
    dIdt = beta*S*I - gamma*I
    dRdt = gamma*I
    return dSdt, dIdt, dRdt
```

```python
# Aplicando o modelo
y0 = S0, I0, R0
t = np.linspace(0,len(dt_filtrado['Data']),len(dt_filtrado['Data']))
ret = odeint(deriv2, y0, t, args=(N, beta, gamma))
S, I, R = ret.T
```

```python
# Representando graficamente os casos acumulados e comparando com a realidade
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], (I+R),'b', lw=2, label = 'Casos acumulados (modelo)')
plt.plot(dt_filtrado['Data'], dt_filtrado['Casos acumulados'],'m', lw=2, label = 'Casos acumulados (realidade)')
plt.legend()
plt.show()
```

```python
# Representando graficamente as curvas de infectados e recuperados
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], (I),'y', lw=2, label = 'Infectados')
plt.plot(dt_filtrado['Data'], (R),'r', lw=2, label = 'Recuperados')
plt.legend()
plt.show()
```

### **Utilizando o Modelo SVIRS**

```python
# Modelo SVIRS

def deriv(y, t, omegaR, omegaV, psi, phi, gamma):
    if t < 219: # até dia 05/02/2021 não havia vacinação - 2ª dose, então todos dados relacionados a ela são 0
      phi, V, omegaV, psi = 0, 0, 0, 0
    if t < 70: # meio de agosto 2020
      r = 0.069
    elif t < 160: # meio de novembro 2020
      r = 0.06
    elif t < 170: # fim de novembro 2020
      r = 0.13
    elif t < 380: # fim de abril 2021
      r = 0.069
    else:
      r = 0.053

    S, V, I, R = y
    beta = r/S
    
    lbda = beta

    dSdt = omegaR*R - lbda*S*I + omegaV*V - phi*S
    dVdt = (psi - 1)*lbda*V*I - omegaV*V + phi*S
    dIdt = lbda*S*I + (1 - psi)*lbda*V*I - gamma*I
    dRdt = gamma*I - omegaR*R

    return dSdt, dVdt, dIdt, dRdt
```

```python
# Aplicando o modelo:

y0 = S0, V0, I0, R0
t = np.linspace(0,len(dt_filtrado['Data']),len(dt_filtrado['Data']))

ret = odeint(deriv, y0, t, args=(omegaR, omegaV, psi, phi, gamma))
S, V, I, R = ret.T
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], (I+R),'b', lw=2, label = 'Casos acumulados')
plt.plot(dt_filtrado['Data'], dt_filtrado['Casos acumulados'],'m', lw=2, label = 'Casos acumulados Verdadeiro')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], (I),'y', lw=2, label = 'Infectados')
plt.plot(dt_filtrado['Data'], (R),'r', lw=2, label = 'Recuperados')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dt_filtrado['Data'], (V),'g', lw=2, label = 'Vacinados (modelo)')
plt.plot(v_graph['Data'], v_graph['Vacinados'],'r', lw=2, label = "Vacinados (realidade)")
plt.legend()
plt.show()
```

Já que utilizamos uma taxa de vacinação constante ao longo do tempo, a curva de vacinados projetada pelo modelo distoa da realidade.

### **Cenário 1: redução das medidas sanitárias**

Nesse cenário, analisaremos que, com a redução de medidas sanitárias e de distanciamento social, maior será a disseminação da doença nos próximos meses. Por isso, consideramos que a infecção aumentará em cerca de 25,5%.

Essa porcentagem é a média de três valores encontrados para o aumento da chance de contaminação considerando o não uso de máscaras. [16]

```python
# Modelo SVIRS

def deriv3(y, t, omegaR, omegaV, psi, phi, gamma):
    if t < 219: # até dia 05/02/2021 não havia vacinação - 2ª dose, então todos dados relacionados a ela são 0
      phi, V, omegaV, psi = 0, 0, 0, 0
    if t < 70: # meio de agosto 2020
      r = 0.069
    elif t < 160: # meio de novembro 2020
      r = 0.06
    elif t < 170: # fim de novembro 2020
      r = 0.13
    elif t < 380: # fim de abril 2021
      r = 0.069
    else:
      r = 0.053*1.255

    S, V, I, R = y
    beta = r/S
    
    lbda = beta

    dSdt = omegaR*R - lbda*S*I + omegaV*V - phi*S
    dVdt = (psi - 1)*lbda*V*I - omegaV*V + phi*S
    dIdt = lbda*S*I + (1 - psi)*lbda*V*I - gamma*I
    dRdt = gamma*I - omegaR*R

    return dSdt, dVdt, dIdt, dRdt
```

```python
# Aplicando o modelo no cenário 1:

y0 = S0, V0, I0, R0
t1 = np.linspace(len(dt_filtrado['Data']),len(dt_filtrado['Data'])+119,119)

ret = odeint(deriv3, y0, t1, args=(omegaR, omegaV, psi, phi, gamma))
S, V, I, R = ret.T
```

```python
# Aplicando novamente o modelo para projeção:

ret2 = odeint(deriv, y0, t1, args=(omegaR, omegaV, psi, phi, gamma))
S2, V2, I2, R2 = ret2.T
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t1,(I+R),'b', lw=2, label = 'Casos acumulados: cenário 1')
plt.plot(t1,(I2+R2),'y', lw=2, label = 'Casos acumulados: projeção')
plt.plot(t1, (I),'r', lw=2, label = 'Infectados')
plt.legend()
plt.show()
```

Dessa forma, percebemos que o uso de máscara continua sendo importante mesmo com a maioria da população vacinada.

### **Cenário 2: surgimento de variante mais infecciosa**

Nesse cenário, consideramos que o surgimento de uma nova variante poderia impactar na taxa de eficácia média das vacinas. Dessa forma, a taxa de vacinação $\psi$ cairá de 76% para 46%.

```python
# Modelo SVIRS

def deriv4(y, t, omegaR, omegaV, psi, phi, gamma):
    if t < 219: # até dia 05/02/2021 não havia vacinação - 2ª dose, então todos dados relacionados a ela são 0
      phi, V, omegaV, psi = 0, 0, 0, 0
    if t < 70: # meio de agosto 2020
      r = 0.069
    elif t < 160: # meio de novembro 2020
      r = 0.06
    elif t < 170: # fim de novembro 2020
      r = 0.13
    elif t < 380: # fim de abril 2021
      r = 0.069
    else:
      r = 0.053
      psi = 0.46

    S, V, I, R = y
    beta = r/S
    
    lbda = beta

    dSdt = omegaR*R - lbda*S*I + omegaV*V - phi*S
    dVdt = (psi - 1)*lbda*V*I - omegaV*V + phi*S
    dIdt = lbda*S*I + (1 - psi)*lbda*V*I - gamma*I
    dRdt = gamma*I - omegaR*R

    return dSdt, dVdt, dIdt, dRdt
```

```python
# Aplicando o modelo no cenário 2:

ret = odeint(deriv4, y0, t1, args=(omegaR, omegaV, psi, phi, gamma))
S, V, I, R = ret.T
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t1,(I+R),'b', lw=2, label = 'Casos acumulados: cenário 2')
plt.plot(t1,(I2+R2),'y', lw=2, label = 'Casos acumulados: projeção')
plt.legend()
plt.show()
```
